```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(42)
N = 1000
vehicle_counts = np.random.choice([0, 1, 2, 3, 4, 5], size=N, p=[0.1, 0.3, 0.3, 0.15, 0.1, 0.05])
unique, counts = np.unique(vehicle_counts, return_counts=
                           True)
pmf = counts / N
cdf = np.cumsum(pmf)
plt.figure(figsize=(10, 4))
sns.barplot(x=unique, y=pmf, palette='viridis', edgecolor='black')
plt.xlabel("Number of Vehicles")
plt.ylabel("Probability")
plt.title("Probability Mass Function (PMF)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
plt.figure(figsize=(10, 4))
plt.step(unique, cdf, where='mid', color='b', label='CDF')
plt.scatter(unique, cdf, color='red')
plt.xlabel("Number of Vehicles")
plt.ylabel("Cumulative Probability")
plt.title("Cumulative Distribution Function (CDF)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom
np.random.seed(42)
N = 1000
vehicle_counts = np.random.choice([0, 1, 2, 3, 4, 5], size=N, p=[0.1, 0.3, 0.3, 0.15, 0.1, 0.05])
unique, counts = np.unique(vehicle_counts, return_counts=True)
pmf = counts / N
cdf = np.cumsum(pmf)
plt.figure(figsize=(10, 4))
sns.barplot(x=unique, y=pmf, palette='viridis', edgecolor='black')
plt.xlabel("Number of Vehicles")
plt.ylabel("Probability")
plt.title("Probability Mass Function (PMF)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
plt.figure(figsize=(10, 4))
plt.step(unique, cdf, where='mid', color='b', label='CDF')
plt.scatter(unique, cdf, color='red')
plt.xlabel("Number of Vehicles")
plt.ylabel("Cumulative Probability")
plt.title("Cumulative Distribution Function (CDF)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
N_gadgets = 100  # Number of gadgets inspected
p_defective = 0.05  # Probability of a gadget being defective
prob_10_defective = binom.pmf(10, N_gadgets, p_defective)
print(f"P(X=10) = {prob_10_defective:.4f}")
prob_at_least_15 = 1 - binom.cdf(14, N_gadgets, p_defective)
print(f"P(X >= 15) = {prob_at_least_15:.4f}")
expected_defective = N_gadgets * p_defective
print(f"Expected defective gadgets: {expected_defective}")
```
